# Finding communities

**Note:** Library `community` can be installed with `conda install -c conda-forge python-louvain`.

```python
import community
import networkx as nx
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from networkx.algorithms.community import greedy_modularity_communities

plt.rcParams["figure.figsize"] = (16, 10)
```

```python
df = pd.read_csv("data/network_metadata.tab", sep="\t", index_col="GexfID")
```

### Good Will Hunting

```python
good_will_hunting = nx.read_gexf("data/gexf/361.gexf", relabel=True)
```

```python
part = community.best_partition(good_will_hunting)
values = [part.get(node) for node in good_will_hunting.nodes()]

nx.draw_spring(good_will_hunting, cmap = plt.get_cmap('prism'), node_color = values, node_size=500, with_labels=True)
```

Louvain's algorithm came up with two communities: the university staff, and Will's friends.
